Veri setini incelediğimiz zaman ortalama saldırı 77 iken maksimum saldırının 16.280 gibi oldukça yüksek seviyede olması verilerde dengesizlik yaratabilir.Uç değerleri çıkartarak örnegin verimizi %95'lik dilime göre temizleyebiliriz

```python
#Veri setinde filtreleme ve dengeleme yapalım
new_data=data['Terrorist attacks'].quantile(0.95)

filter_data=data[data['Terrorist attacks']<=new_data]

#Yıllara göre saldırıları alalım
terorist_attack=filter_data.groupby('Year')['Terrorist attacks'].sum()

#matplotlib import
import matplotlib.pyplot as plt


#Görselleştirme
plt.figure(figsize=(10,6))
plt.plot(terorist_attack.index,terorist_attack.values,marker="o",linestyle="-")
plt.title('Yıllara göre terörist saldırılar')
plt.xlabel('Yıl')
plt.ylabel('Toplam Terörist Saldırılar')
plt.grid(True)
```

YILLARA GÖRE TERÖR SALDIRILARI 

Bu not,terör saldırılarının yıllara göre bir zaman serisinde incelemek için hazırlanmıştır.Bazı verilerde aşırı uç değerler gözlemlendiği için verilerin temizlenmesi gerekmiş olup terör saldırılarının zamansal değişimini ortaya koymayı amaçlamaktadır

```python
#En çok saldırı yaşayan ülke ve kayıp veren ülkeler listesi

country_attack=data.groupby('Entity')["Terrorist attacks"].sum()
country_deaths=data.groupby('Entity')['Terrorism deaths'].sum()

#Aykırı değer temizliği
attack_filter=country_attack.quantile(0.95)
death_filter=country_attack.quantile(0.95)

#Filtreleme yapalım
filter_country_attack=country_attack[country_attack<=attack_filter]
filter_country_death=country_deaths[country_deaths<=death_filter]

#Sıralama ile ilk 5 ülkeye bakalım
top_attack=filter_country_attack.sort_values(ascending=False).head(5)
top_death=filter_country_death.sort_values(ascending=False).head(5)


fig, axs = plt.subplots(2, 1, figsize=(12, 10))


# Top5 saldırı
axs[0].barh(top_attack.index, top_attack.values)
axs[0].set_title('En çok saldırı yaşayan 5 Ülke)')
axs[0].set_xlabel('Toplam Terörist Saldırı')
axs[0].invert_yaxis()

# Top5 ölüm
axs[1].barh(top_death.index, top_death.values, color='orange')
axs[1].set_title('En çok kayıp yaşayan 5 ülke')
axs[1].set_xlabel('Teröre Bağlı Ölüm')
axs[1].invert_yaxis()

```

```python

import pandas as pd
import matplotlib.pyplot as plt



# Tekrar verileri Türkiye'ye göre filtreleyelim
turkey_data = data[data['Entity'] == 'Turkey']


# Filtreleme işlemleri
saldırı_metot = [col for col in turkey_data.columns if 'Attack method' in col]
turkey_saldırı_metot = turkey_data[saldırı_metot]

# Toplamlarını alalım
toplam_saldırı=  turkey_saldırı_metot.sum().sort_values(ascending=False)

# Plotting the attack methods distribution
plt.figure(figsize=(12, 6))
plt.bar(toplam_saldırı.index, toplam_saldırı.values)
plt.title('Distribution of Terrorist Attack Methods in Turkey')
plt.xlabel('Attack Methods')
plt.ylabel('Total Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
